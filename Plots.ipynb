```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.subplots()
ax.plot([1, 2, 3, 4], [0, 0.5, 1, 0.2])
```

```python
fig, ax = plt.subplots(1,2)
ax[0].plot([1, 2, 3, 4], [0, 0.5, 1, 0.2])
ax[1].plot([1, 2, 3, 4], [-0.1, 0.5, -1, 0.8])
```
